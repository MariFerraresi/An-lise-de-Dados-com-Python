Quando entramos com este comando, o python exibe somente as linhas onde faltam os dados. É necessário entender a importância do dado para nosso projeto para definir qual será o destino do dado.

Vejamos agora como excluir uma linha dentro do dataframe

```python
compras = compras.drop(45)
```

```python
compras.drop(34, inplace= True)
```

```python
compras
```

```python
compras[compras.isnull().any(axis=1)]
```

Neste caso acima observa-se que o problema da coluna com dado nulo é um problema de operação, e assim , podemos realizar a operação e preencher o dado faltante

```python
compras.loc[33, "Total"] = compras.loc[33, "Unidade"] *compras.loc[33,"Preço Unidade"]
```

```python
compras
```

```python
compras.loc[40, "Total"] = compras.loc[40, "Unidade"] *compras.loc[40,"Preço Unidade"]
```

```python
compras
```

```python
compras.loc[42, "Preço Unidade"] = compras.loc[42, "Total"] /compras.loc[42,"Unidade"]
```

```python
compras
```

```python
compras[compras.isnull().any(axis=1)]
```

diferença na formatação da data que pode dar problema posterior

```python
compras.Data = compras.Data.str.replace("-","/")
```

```python
compras.Data = compras.Data.str.replace("//" , "/")
```

```python
compras
```

```python
preco_ajustado =[]
for preco in compras ["Preço Unidade"]:
    preco_ajustado.append(float(str(preco).replace(";" , ".")))
```

```python
compras["Preço Unidade"] = preco_ajustado
```
